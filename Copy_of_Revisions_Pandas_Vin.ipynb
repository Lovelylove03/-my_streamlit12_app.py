```python
import pandas as pd
from sklearn.impute import KNNImputer

# Separate numerical and categorical columns
categorical_cols = df_cleaned.select_dtypes(include=['object']).columns
numerical_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns

# Handle categorical columns - fill missing values with the mode (most frequent value)
for col in categorical_cols:
    df_cleaned[col].fillna(df_cleaned[col].mode()[0], inplace=True)

# Now apply KNN Imputer only on the numerical columns
imputer = KNNImputer(n_neighbors=5)
df_cleaned[numerical_cols] = imputer.fit_transform(df_cleaned[numerical_cols])

# After imputation, check the DataFrame to ensure everything is correct
print(df_cleaned.info())
```

```python
df_cleaned.isnull().sum()
```

```python
nan_percentage = (df_cleaned.isna().mean() * 100)
nan_percentage = nan_percentage.sort_values(ascending=False)
nan_percentage
```

```python
# Step 4: Simple imputation for columns with ~5% missing values
columns_with_low_missing = ['post_modified_gmt', 'post_modified', 'post_date_gmt', 'post_date']

# Impute missing values with the mode for categorical and median for numerical
for col in columns_with_low_missing:
    if df_cleaned[col].dtype == 'object':  # Categorical
        df_cleaned[col].fillna(df_cleaned[col].mode()[0], inplace=True)
    else:  # Numerical
        df_cleaned[col].fillna(df_cleaned[col].median(), inplace=True)

# Step 5: Ensure there are no remaining NaN values
df_cleaned = df_cleaned.dropna()
```
